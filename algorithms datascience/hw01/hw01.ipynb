```python
n = 0
with open(path) as f:                   # The file object is named f.
    for string in f:                    # Process each record in the file.
        data = string.split("|")
```

```python
indivDict[data[7]] += int(data[14])
```

```python
with open(path) as f:
    for string in f:
        data = string.split("|")
        indivDict[data[7]] += int(data[14])
```

```python
n += 1
if n % 5000 == 0:
    print(n)
```

```python
print(len(indivDict))
```

```python
sortedSums = sorted(indivDict.items(), key=operator.itemgetter(1))
```

```python
for item in sortedSums :
    if item[1] >= 25000 : print(item[1], item[0])
```

```python
for key, value in sortedSums[-10:]:
    print(f"{key}: {value}")
```

<h1>The way I work on the data</h1>

```python
column_names = [
    'Committee_ID',           # C00403477 -  
    'Amendment_Indicator',    # N -  
    'Report_Type',            # 12S -  
    'Unused',                 #   ()
    'Filing_Number',          # 13961108044 -  
    'Number',                 # 15 -    
    'Entity_Type',            # IND -  
    'Contributor_Name',       # HENCHBARGER, ROBERT -  
    'City',                   # CENTREVILLE - 
    'State',                  # VA - 
    'ZIP_Code',               # 20120 - 
    'Employer',               # MICHAEL BAKER JR -  
    'Occupation',             # OPERATIONS MANAGER V -  
    'Transaction_Date',       # 02222013 -  
    'Transaction_Amount',     # 220 -  
    'Other_ID',               #  
    'Report_ID',              # SA11AI.6023 -  
    'FEC_ID',                 # 858916 -  FEC
    'Empty1',                 #  
    'Empty2',                 #  
    'Image_Number'            # 4030420131184956987 -  
]
```

```python
df = pd.read_csv(path, sep='|', names=column_names, engine='python')

df.head()
```

```python
contributor_summary = df.groupby(['Contributor_Name', 'ZIP_Code']).agg(
                                                                        Total_Amount=('Transaction_Amount', 'sum'),  
                                                                        Count_Amount=('Transaction_Amount', 'size')       
                                                                        ).reset_index()

contributor_summary = contributor_summary.sort_values(by='Total_Amount', ascending=False).reset_index(drop=True)

contributor_summary.head(10)
```

````python
q = 3
A = np.zeros(shape = (q, q))
z = np.matrix(np.zeros(shape = (q, 1)))
sumSqrs = 0
n = 0
````

````python
path = r'Algorithms/Data/'
fileList = os.listdir(path)
for filename in fileList:
    shortYear = int(filename[6:8])
    if shortYear < 11:
        1/0
    year = 2000 + shortYear
    file = path + filename
    fields = fieldDict[shortYear]
    
    sInc, eInc = fields['income']
    sBMI, eBMI = fields['bmi']
    fEduc = fields['education']
    fGH = fields['genhlth']
    
    print(file,sInc, eInc,sBMI, eBMI, fEduc, fGH)

    with open(file, encoding = "latin-1") as f:
        for record in f:
            education = getEducation(record[fEduc-1])
            
            incomeString = record[sInc-1:eInc]
            try:
                income = getIncome(incomeString)
            except ValueError:
                income = 9
                
            bmiString = record[sBMI-1:eBMI]
            try:
                bmi = convertBMI(bmiString, shortYear)
            except ValueError:
                bmi = 0
                
            y = getHlth(record[fGH-1])
            
            try:
                if education < 9 and income < 9 and 0 < bmi < 99 and y != -1:                      
                    x = np.matrix([1, income, bmi]).T
                    A += x*x.T
                    z += x*y
                    sumSqrs += y**2
                    n += 1
                    if n % 10000 == 0 and n != 0:
                        b = np.linalg.solve(A,z)
                        ybar = z[0]/n
                        varEst = sumSqrs/n - ybar**2
                        rAdj = (z.T.dot(b)/n - ybar**2)/varEst
                        bList_no_educ = [str(round(float(bi),3)) for bi in b]
                else:
                    1/0
            except(ZeroDivisionError):
                pass
````
